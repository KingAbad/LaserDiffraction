```python
import numpy as np
import matplotlib.pyplot as plt
import xlsxwriter


workbook = xlsxwriter.Workbook('E:\Stage\PDF.xlsx')
worksheet = workbook.add_worksheet()
workbook1 = xlsxwriter.Workbook('E:\Stage\X.xlsx')
worksheet1 = workbook1.add_worksheet()
  
# Start from the first cell.
# Rows and columns are zero indexed.
row = 0
column = 0
row1 = 0
column1 = 0

mu = 20000e-9 #Mean - droplet diameter in nm
sigma = 4000e-9 #Standard deviation

s = np.random.lognormal(mu, sigma, 100000)


count, bins, ignored = plt.hist(s, 1000, density=True, align='mid')

x = np.linspace(min(bins), max(bins), 1000)

pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))/ (x * sigma * np.sqrt(2 * np.pi)))

plt.plot(x, pdf, linewidth=2, color='r')

plt.axis('tight')

plt.show()

row = 0

# iterating through content list
for item in pdf :
  
    # write operation perform
    worksheet.write(row, column, item)
  
    # incrementing the value of row by one
    # with each iteratons.
    row += 1
      
workbook.close()

row1 = 0

# iterating through content list
for item in x :
  
    # write operation perform
    worksheet1.write(row1, column1, item)
  
    # incrementing the value of row by one
    # with each iteratons.
    row1 += 1
    
workbook1.close()

print(x)
print(pdf)
```
